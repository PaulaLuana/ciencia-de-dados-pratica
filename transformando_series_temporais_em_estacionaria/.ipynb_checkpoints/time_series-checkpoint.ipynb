# Série temporal estacionária.

Para construir um modelo de série temporal é necessário que a série seja estacionária. Em casos que a série não é estacionária, a primeira coisa que devemos fazer é torna-la estacionária, em seguida, tentar modelos estocásticos para prever essa série temporal. Existem várias maneiras de trazer essa estacionariedade e é o que vamos aprender hoje.

## Mas o que é uma série estacionária?

Uma série é estacionária quando suas propriedades estatísticas permanecem constantes. Mais especificamente, uma série é estacionária se satisfazer as seguintes condições.

1. A média da série não deve ser uma função do tempo, mas sim uma constante. A imagem abaixo tem o gráfico do lado esquerdo satisfazendo a condição, enquanto o gráfico em vermelho tem uma média dependente do tempo.

![Imagem do site Analytics Vidhya](img/grafico-1.png)

2. A variação da série não deve ser função do tempo. Esta propriedade é conhecida como homocedasticidade. O gráfico a seguir mostra o que é e o que não é uma série estacionária. (Observe a variação da distribuição no gráfico à direita)

![Imagem do site Analytics Vidhya](img/grafico-2.png)

3. A covariância do i ésimo termo e do (i + m) ésimo termo não deve ser função do tempo. No gráfico a seguir, você notará que o spread se torna mais próximo à medida que o tempo aumenta. Portanto, a covariância não é constante com o tempo para a 'série vermelha'.

![Imagem do site Analytics Vidhya](img/grafico-3.png)

### Aplicando teste de estacionaridade

Para o teste de estacionaridade vamos usar o teste de Dickey-Fuller, em que já temos a função adfuller na biblioteca statsmodels. Observe que na função que definimos a seguir usamos a média móvel e o desvio padrão móvel para visualizar o comportamento dos nossos dados e logo após, realizamos o teste.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean() #media movel
    rolstd = timeseries.rolling(12).std() #desvio movel

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Média móvel & desvio padrão')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Resultados do teste de Dickey-Fuller')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estatística do teste','p-value','O critério de informação maximizado ','Número de observações usadas'])
    for key,value in dftest[4].items():
        dfoutput['Valor crítico (%s)'%key] = value
    print(dfoutput)
```

**Teste de Dickey-Fuller:** Este é um teste do tipo Hipótese de Raiz Unitária no qual ele tenta determinar determinar o quão forte a série temporal é definida por uma tendência.

Basicamente vamos ter uma tabela que apresentará o que chamamos de valor critico, que é uma espécie de valor de corte. Em nosso exemplo são exibidos níveis de 1%, 5% e 10%. Se o resultado do ADF for menor do que o valor critico para um determinado percentual (5% no nosso caso) e o p-value for significante (menor que 0.05), a série é considerada estacionária. Do contrário a série é não estacionária.



Com as condições que temos sobre uma série ser estacionária, vamos criar uma função simples que verifica isso para gente.

```python
def ehEstacionaria(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estatística do teste','p-value','O critério de informação maximizado ','Número de observações usadas'])
    for key,value in dftest[4].items():
        dfoutput['Valor crítico (%s)'%key] = value
    
    if(dfoutput['Estatística do teste'] < dfoutput['Valor crítico (5%)'] and dfoutput['p-value'] < 0.05):
        print('É estacionária.')
    else:
        print('Não é estacionária.')
```

```python
test_stationarity(df)
```

```python
ehEstacionaria(df)
```

Como podemos ver essa série não é estacionária ainda. Pensando num nível de significância de 5%, o p-value e o valor do teste estão muito altos.

## Tornando a série estacionária

As técnicas usadas para conseguir uma série estacionária, são, transformações (logarítmica, a raiz quadrada, boxcox), diferenciação, etc.<br>
Vamos começar com o método da diferenciação. Esse processo cria uma nova série a partir da diferença de uma observação com a observação anterior. 

Através da diferenciação também é possível remover a dependência temporal de uma série, reduzindo a tendência e a sazonalidade. Por esse motivo esse método pode ser usado para atingir a estacionariedade. Pode-se diferenciar uma série mais de uma vez, caso seja necessário para remover a dependência.

```python
dfdiff = df.daily_vaccinations.diff()
dfdiff = dfdiff.dropna()
plt.title("Primeira diferenciação")
dfdiff.plot()
plt.show()
```

**A transformação BoxCox** tem como objetivo deixar os dados mais normalizados e consequentemente estabiliza a variância durante os dados. A sua função tem apenas um parâmetro, lambda, que denota como será feita essa transformação, podemos definir esse valor ou deixar a função achar “o melhor valor”

```python
from scipy.stats import boxcox
meantemp_bcx, lam = boxcox(df['daily_vaccinations'])
print('Lambda: %f' % lam)

fig, ax = plt.subplots()
# line plot
fig.suptitle("BoxCox resultados")
plt.subplot(211)

plt.plot(meantemp_bcx)
# histogram
plt.subplot(212)
plt.hist(meantemp_bcx)
plt.show()
```

```python
print('Results of Dickey Fuller Test:')
dftest = adfuller(meantemp_bcx, autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

```python
ehEstacionaria(meantemp_bcx)
```
